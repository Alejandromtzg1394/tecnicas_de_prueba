## Cobertura con Python

## Ejecución para métricas de cobertura

```python

Sintaxis:

pytest -sv --cov=[directorio_fuentes] [directorio_tests] 



Ejemplo:

pytest -sv --cov=app_web tests



donde:

    app_web - directorio de los archivos fuente, para identificar la cobertura
    tests - directorio donde se encuentran los tests 


```

## Ejecución con generación del reporte en HTML

```python

Sintaxis:

pytest -sv --cov=[directorio_fuentes] --cov-report=[FORMATO]:[directorio_salida] [directorio_tests] 



Ejemplo:

pytest -sv --cov=app_web --cov-report=html:reporte tests




donde:

    app_web  - directorio de los archivos fuente, para identificar la cobertura
    tests    - directorio donde se encuentran los tests 
    html     - formato de salida
    reporte  - directorio de salida para los archivos generados


```

````python
class PersonaValueError(Exception):
    pass

class Persona:
    """Clasifica a una persona en las diferentes etapas \
    de la vida según su edad."""
    def __init__(self, edad):
        if edad < 0:
            raise PersonaValueError("La edad no puede ser negativa.")
        self.edad = edad
    
    def clasificar_etapa_vida(self):
        if self.edad <= 12:
            return "Infancia"
        elif 13 <= self.edad <= 19:
            return "Adolescencia"
        elif 20 <= self.edad <= 64:
            return "Adultez"
        else:
            return "Vejez"
            
# Ejemplo de uso
p1 = Persona(13)
# debe retornar "Adolescencia"
print(p1.clasificar_etapa_vida())

# lanza una excepción "PersonaValueError"
p1 = Persona(-10)
print(p1.clasificar_etapa_vida())    
````
